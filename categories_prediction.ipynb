```python
import pandas as pd
books = pd.read_csv("./cleaned_books.csv")
```

```python
import numpy as np
books["categories"].value_counts().reset_index().query("count > 0")[:60]
```

```python
category_mapping = {
    "Fiction":"Fiction",
    "Juvenile Fiction":"Children's Fiction",
    "Biography & Autobiography": "Nonfiction",
    "History":"Nonfiction",
    "Literary Criticism":"Nonfiction",
    "Philosophy":"Nonfiction",
    "Religion":"Nonfiction",
    "Comics & Graphic Novels":"Fiction",
    "Juvenile Nonfiction":"Children's Nonfiction",
    "Science":"Nonfiction",
    "Poetry":"Fiction",
    "Computers":"Nonfiction",
    "Business & Economics": "Nonfiction",
    "Literary Collections":"Fiction",
    "Cooking":"Nonfiction",
    "Children's stories":"Children's Fiction",
    "Adventure stories":"Fiction",
    "Detective and mystery stories":"Fiction",
    "Fantasy fiction":"Fiction",
    "English fiction": "Fiction",
    "American fiction": "Fiction",
    "Science fiction": "Fiction",
    "Young Adult Fiction": "Fiction",
    "FICTION": "Fiction",
}
```

```python
books["Maped_categories"] = books["categories"].map(category_mapping)
```

```python
books[~(books["Maped_categories"].isna())]
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

categories = ["Fiction", "Nonfiction"]

pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device="cuda")
```

```python
sequence = books.loc[books["categories"] == "Fiction", 'description'].reset_index(drop=True)[0]
pipe(sequence, categories)
```

```python
def get_prediction(sequence, categories):
    prediction = pipe(sequence, categories)
    max_index = np.argmax(prediction["scores"])
    max_label = prediction["labels"][max_index]
    return max_label
```

```python
from tqdm import tqdm
actual_cats = []
predicted_cats = []
for i in tqdm(range(0, 300)):
    sequence = books.loc[books["Maped_categories"] == "Fiction", 'description'].reset_index(drop=True)[i]
    predicted_cats += [get_prediction(sequence, ["Fiction", "Nonfiction"])]
    actual_cats += ["Fiction"]
```

```python
for i in tqdm(range(0, 300)):
    sequence = books.loc[books["Maped_categories"] == "Nonfiction", 'description'].reset_index(drop=True)[i]
    predicted_cats += [get_prediction(sequence, ["Fiction", "Nonfiction"])]
    actual_cats += ["Nonfiction"]
```

```python
predicted_df = pd.DataFrame({"actual_cat":actual_cats, "predicted_cat":predicted_cats})
```

```python
predicted_df["correct_pred"] = (np.where(predicted_df["actual_cat"] == predicted_df["predicted_cat"], 1, 0))
```

```python
predicted_df["correct_pred"].sum() / len(predicted_df)
```

```python
missing_cat = books.loc[books["Maped_categories"].isna(), ['isbn13','description']].reset_index(drop=True)
```

```python
isbns = []
prdicted_categories = []
for i in tqdm(range(0, len(missing_cat))):
    sequence = missing_cat['description'][i]
    prdicted_categories += [get_prediction(sequence, ["Fiction", "Nonfiction"])]
    isbns += [missing_cat['isbn13'][i]]
```

```python
missing_predicted_df = pd.DataFrame({"isbn13":isbns, "prdicted_categories":prdicted_categories})
missing_predicted_df
```

```python
books = pd.merge(books, missing_predicted_df, on="isbn13", how="left")
books["Maped_categories"] = np.where(books["Maped_categories"].isna(), books["prdicted_categories"], books["Maped_categories"])
books = books.drop(columns = ["prdicted_categories"], axis=1)
```
